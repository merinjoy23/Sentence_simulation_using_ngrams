```python
import re, random , sys , string, fileinput
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.util import ngrams
    
f1 = input('Enter first file: ')
f2 = input('Enter second file: ')
f3 = input('Enter third file: ')

f11 = open(f1, encoding ="utf-8")
f22 = open(f2, encoding ="utf-8")
f33 = open(f3, encoding ="utf-8")

f1_contents = f11.read()
f2_contents = f22.read()
f3_contents = f33.read()

f4 = open("new.txt","w", encoding ="utf-8")
f4.write(f1_contents+f2_contents+f3_contents)
f4.close()

f5 = open("new.txt", encoding="utf-8") 
text = f5.read()

print (len(text))
```

```python
words = text.lower()
tokens1 = nltk.word_tokenize(words)
tokens1
print(len(tokens1))
```

```python
gram1 = dict()
# Populate 1-gram dictionary
for word in tokens1:
    if word in gram1.keys():
        gram1[word] += 1
    else:
        gram1[word] = 1 # Start a new entry with 1 count since saw it for the first time.

# Turn into a list of (word, count) sorted by count from most to least
gram1 = sorted(gram1.items(), key= lambda x:-x[1])

# Print top 20 most frequent words
print (gram1)
```

```python
gram2 = dict()

# Populate 2-gram dictionary
for i in range(len(tokens1)-1):
    key = (tokens1[i], tokens1[i+1])
    if key in gram2.keys():
        gram2[key] += 1
    else:
        gram2[key] = 1

# Turn into a list of (word, count) sorted by count from most to least
gram2 = sorted(gram2.items(), key= lambda x:-x[1])

# Print top 20 most frequent words
print (gram2)
```

```python
import random
def weighted_choice(choices):
   total = sum(w for c, w in choices)
   r = random.uniform(0, total)
   upto = 0
   for c, w in choices:
      if upto + w > r:
         return c
      upto += w

def get2GramSentenceRandom(word, n = 200):
    for i in range(n):
        print (word, end = " ")
        # Get all possible elements ((first word, second word), frequency)
        choices = [element for element in gram2 if element[0][0] == word]
        if not choices:
            break

        # Choose a pair with weighted probability from the choice list
        word = weighted_choice(choices)[1]
        
    for word in ['and', 'he', 'she', 'when', 'john', 'never', 'i', 'how']:
        print ("Start word: %s" %word)

print ("2-gram sentence: \"")
get2GramSentenceRandom(word)
print ("\"")
```

```python
def generateNgram(n=1):
    gram = dict()

    # Some helpers to keep us crashing the PC for now
    assert n > 0 and n < 100

    # Populate N-gram dictionary
    for i in range(len(tokens1)-(n-1)):
        key = tuple(tokens1[i:i+n])
        if key in gram.keys():
            gram[key] += 1
        else:
            gram[key] = 1

    # Turn into a list of (word, count) sorted by count from most to least
    gram = sorted(gram.items(), key=lambda x: -x[1])
    return gram

trigram = generateNgram(3)
# Print top 20 most frequent ngrams
print (trigram)
```
